```python
import numpy as np
import h5py
import pandas as pd
import matplotlib.pyplot as plt
from statistics import pstdev
```

```python
h5 = h5py.File('data/SG24_dataset.h5')
```

```python
print(list(h5.keys()))
print(h5['Predictors'])
print(h5['Target'])
print(h5['User'])
```

```python
predictors = np.asarray(h5['Predictors'])
pred = pd.DataFrame(predictors)
print(pred)

target = np.asarray(h5['Target'])
tar = pd.DataFrame(target[0])
print(tar)

users = np.asarray(h5['User'])
user = pd.DataFrame(users[0])
print(user)
```

```python
dico_users=dict()
for user in users[0]:
    if user in dico_users.keys():
        dico_users[user]+=1
    else :
        dico_users[user]=1
        

plt.bar(dico_users.keys(), dico_users.values())
plt.xlabel("Sujet")
plt.ylabel("Nombre de gestes")
```

```python
dico_target=dict()
for label in target[0]:
    if label in dico_target.keys():
        dico_target[label]+=1
    else :
        dico_target[label]=1
        
plt.bar(dico_target.keys(), dico_target.values())
```

```python
for i in range(1,9):
    print("user : "+str(i))
    l=np.where(np.array(users[0]) == i)[0]
    dico_target=dict()
    for j in l:
        label=target[0][j]
        if label in dico_target.keys():
            dico_target[label]+=1
        else :
            dico_target[label]=1
    plt.pie(list(dico_target.values()),labels = dico_target.keys())
    plt.title("proportion des labels pour le user "+str(i))
    plt.show() 
    plt.bar(dico_target.keys(), dico_target.values())
    plt.show() 
    
```
